```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#inicializaci√≥n
graph = nx.Graph()
```

```python
people = ['jere','ella','anniina','mikka','mikko','olli','laura','maria']
```

```python
connections = [('jere','ella'),('ella','anniina'),('ella','mikka'),('mikko','ella'),('anniina','mikko'),('laura','jere'),
              ('olli','jere'),('jere','maria'),('mikka','mikko'),('maria','laura'),('olli','laura')]
```

```python
#agrego todos los nodos
for node in people:
    graph.add_node(node)
```

```python
#agrego conecciones
for node_a,node_b in connections:
    graph.add_edge(node_a,node_b)
```

```python
plt.figure(figsize=(25, 8))
nx.draw_networkx(graph, node_size = 2300)
```

```python
#agregar color por genero
color_hombre = 'green'
color_mujer = 'orange'

color_by_genero = { 'jere' : color_hombre,'ella' : color_mujer,'anniina' : color_hombre,'mikka' : color_hombre
                   ,'mikko' : color_hombre,'olli' : color_hombre,'laura' : color_mujer,'maria' : color_mujer}

#colores por nodo
colors = [color_by_genero[node] for node in graph.nodes()]

#dibujo el grafo
plt.figure(figsize=(25, 8))
nx.draw_networkx(graph, node_size=2300, node_color = colors)

```

```python
# AGREGO COLUMNAS DE GRAFICOS (sin color, y con color), PARA DIFERENCIARLOS, Y DEJO FIJO LOS NODOS.
#dejo fijo los nodos
pos = nx.layout.spring_layout(graph)

fig, ax = plt.subplots(2, 1, figsize=(25, 14))

left_plot = ax[0]
right_plot = ax[1]

nx.draw_networkx(graph, pos = pos,ax=left_plot,node_size =2000)
nx.draw_networkx(graph, pos = pos,ax=right_plot, node_size=2000, node_color = colors)
```

```python
#veamos el grado de conectividad de cada nodo
numero_de_conexiones = graph.degree()

for person in numero_de_conexiones:
    print("{} tiene {} amigos/conexiones".format(person[0],person[1]))
```

```python
#agregar color por genero
color_hombre = 'green'
color_mujer = 'orange'

color_by_genero = { 'jere' : color_hombre,'ella' : color_mujer,'anniina' : color_hombre,'mikka' : color_hombre
                   ,'mikko' : color_hombre,'olli' : color_hombre,'laura' : color_mujer,'maria' : color_mujer}

#colores por nodo
colors = [color_by_genero[node] for node in graph.nodes()]

#degree
dg = nx.degree(graph)
degree_persons = [dg[node] * 800 for node in graph.nodes()]

#dibujo el grafo
plt.figure(figsize=(25, 8))
nx.draw_networkx(graph, node_size=degree_persons, node_color = colors)

```
